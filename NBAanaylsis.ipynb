###  1. Import libraries and load data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('nbaplayersdraft.csv')
df.head(10)
```

### 2. Data Preprocessing

```python
# Check for missing values
print(df.isnull().sum())

#Drop people who have more then 3 columns of missing data
df = df.dropna(thresh=len(df.columns) - 3)

#From the 1980s, early 2000s, some data will be not avaialble.



```

```python
df.info()
#All data types are correct
```

```python
df.describe()
```

### Total points scored by overall picks and years

```python
df = df[df["games"] > 245] #Ensure players have played 3 seasons worth of games

fig = px.scatter_3d(df, x = "year", y = "overall_pick" , z = "points",
                    opacity = 0.9, hover_data = ["player"],
                    color = "overall_pick", color_continuous_scale = "RdBu")

fig.update_traces(marker = dict(size = 3)) #make markers smaller
fig.update_layout(template="plotly_dark")

fig.show()

# Can see strong correlation of higher overall picks amounting to greater points in their career
# Can see players like lebron james who are clear above the rest
```

```python
df = df[df["games"] > 245] #Ensure players have played 3 seasons worth of games

fig = px.scatter_3d(df, x = "year", y = "overall_pick" , z = "assists",
                    opacity = 0.9, hover_data = ["player"],
                    color = "overall_pick", color_continuous_scale = "RdBu")

fig.update_traces(marker = dict(size = 10)) #make markers smaller
fig.update_layout(template="plotly_dark")

fig.show()

#Here the correlation can still be seen, but more lower draft picks amount to higher total assists, such as steve nash and rondo
```

```python
df = df[df["games"] > 245] #Ensure players have played 3 seasons worth of games

fig = px.scatter_3d(df, x = "year", y = "overall_pick" , z = "total_rebounds",
                    opacity = 0.9, hover_data = ["player"],
                    color = "overall_pick", color_continuous_scale = "RdBu")

fig.update_traces(marker = dict(size = 3)) #make markers smaller
fig.update_layout(template="plotly_dark")

fig.show()

#Again there is less correlation here, with more lower drafts picks amounting to high total rebounds
```

```python
df = df[df["games"] > 245] #Ensure players have played 3 seasons worth of games

fig = px.scatter_3d(df, x = "year", y = "overall_pick" , z = "win_shares",
                    opacity = 0.9, hover_data = ["player"],
                    color = "overall_pick", color_continuous_scale = "RdBu")

fig.update_traces(marker = dict(size = 3)) #make markers smaller
fig.update_layout(template="plotly_dark")

fig.show()

#Win shares closer resemble the total points category, so is it evident to say more points contribute to greater winning ability
```

```python
df = df[df["games"] > 250]

fig = px.scatter_3d(
    df, x = "year",
    y = "overall_pick", 
    z = "value_over_replacement", 
    opacity = 0.8, 
    hover_data = ["player"],
    color = "overall_pick", 
    color_continuous_scale = "RdBu")

fig.update_traces(marker = dict(size = 3.5)) # scaling down the markers
fig.update_layout(template = "plotly_dark")
fig.show()

#Similary results to total win shares 
```

```python
df = df[df["games"] > 250]

fig = px.scatter(
    df,
    x='points_per_game', 
    y='value_over_replacement', 
    color='value_over_replacement',  # Use VORP for coloring
    color_continuous_scale='rdbu', 
    opacity = 0.8,
    hover_name='player',  # Displays player names on hover
    trendline='ols',  # Adds a trendline to show the correlation
    title='Correlation between Points Per Game and Value Over Replacement',
)

fig.update_layout(template = "plotly_dark")



fig.show()

# As you can see there is a fairly positive correlation between PPG and VORP
# There are outliers, that are not even getting above 20PPG, and having high VORP, These are players such as Kevin garnet, chris paul, Tim duncan
# These are players who can impact the game in other ways, such as rebounding , assists and defending
```

### Anaylsing draft picks

```python
for y in ["minutes_played", "games", "years_active", "points",
            "win_shares", "value_over_replacement"]:
    
    fig = px.bar(df, x= "overall_pick", 
    y = y, 
    title = f"NBA draft overall pick 1-60 Total : {y}",
    color = "overall_pick",
    color_continuous_scale = 'agsunset')

    fig.update_layout(template = "plotly_dark")

    fig.show()
```

## Insights 

- We can see the top 5 greatly outperform the rest of the top 10 in all metrics

- The 3,4,5 picks tend to do better then the 2nd overall pick

- The 7th, 8th pick are historically the worse picks out of the top 10
